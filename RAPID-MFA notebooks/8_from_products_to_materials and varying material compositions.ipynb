# RAPID-MFA-8: Switching from products/services to materials with changing material compositions

## Descriptions

This Dynamic MFA model makes the switch from modeling the flows and stocks of a __product (in numbers of units)__ or a __service (in appropriate units of service, e.g. m<sup>2</sup> of floor area, GW of electricity etc.)__ to the flows and stocks of a __material (in mass)__ contained in that product or service.

This model goes by many names:
- The [Demand-driven model](https://doi.org/10.1111/jiec.12284)
- Services-products-materials cascade
- [Müller 2006](https://doi.org/10.1016/j.ecolecon.2005.09.025)-style model (perhaps the one that introduced it in its common form)
- [Multi-layer dynamic model](https://doi.org/10.1111/jiec.13023) depicting products, materials, and so on.
- Stocks-flows-services nexus model
- etc.

It's built of __two daisy-chained dynamic MFA models:__ one for the product/service layer, and one for the materials layer. The **conversion** uses **_material intensities_** of mass per one unit of product or service provisioning.

The materials-layer MFA model __must be a flow driven model__ to ensure that if material intensities vary over time then the materials-level stocks and outflows are correct, accounting for the varying material intensities of each cohort. This means that the switch from products to materials is done through the inflows of both layers:

$
\textcolor{#2A8596}{\text{material inflows}(t)[mass]} = \textcolor{#C58D01}{\text{product inflows}(t)[units]} \times \textcolor{#C32D2E}{\text{material intensities}(t)[\frac{mass}{unit}]}
$

### Notes

1. The example used in this notebook showcases why the switch must be done through the inflows: _material intensities_ have a fixed value for the first ten timeframes (the years 2015-2025) and then start to decline. This is a case of _improving technology / lightweighting_: new product inflows use less and less materials per unit of service.

1. The end of this notebook shows how the change in newer inflows' material compositions gradually affects the _material stock productivity_ of service provisioning.

1. The materials layer should remain a _flow driven model_ for the reasons described above, but the products layer can be modeled by either a _stock driven model_ as in this example, or by a _flow driven model_ by simply replacing the relevant code cell.

1. This daisy-chained model can be extended to further layers beyond the products layer-to-materials layer using the same modeling concepts, switching using layers-conversion coefficients like the _material intensities_ in this example. For instance population-to-service with per-capita service units coefficients, service-to-product using product usage coefficients (e.g. vehicle-km), materials-to-elements, and so on.

1. If more than one material in a product is to be modeled, each material should get its own materials-layer dynamic MFA system.

1. In cases in which material intensities do not change over time and are not expected to change, the conversion from the products layer to the materials layer can be reduced to simply multiplying all product-layer flows and stocks by the fixed-valued material intensity coefficient.

1. In this example, the material is assumed to last as long as the product it's in, and therefore use the same survival curve. Modifying it to use separate survival behavior for the materials layer (e.g. periodic replacement of components during the use period) is beyond the scope of this example.

### The exogenous input variables used in this example
he example uses offshore wind turbines in units of generation capacity (MW) and converted to mass of Neodymium, from [Fishman and Graedel 2019](https://doi.org/10.1038/s41893-019-0252-z):

__Stocks:__ Historical data and nationwide future "central scenario", simplified and modified for this example.

__Survival curve:__ Normal distribution, $mean=20$ from the study, $SD=(mean/2)$ modified for this example.

__Material intensities:__ from the study, adapted from references listed in the study.

## 1. Load libraries

```python
# %% 1. Load libraries
import numpy as np
import pandas as pd
import scipy.stats
import seaborn as sns
```

## 2. Define example input data

```python
# %% 2. Define example input data

input_data = pd.Series([0, 30, 30, 30, 30, 30, 2985, 6523, 8721,
                        10171, 11851, 13772, 15729, 17621, 19480, 21761, 24793, 27959,
                        30680, 33460, 36803, 40144, 42956, 45948, 49759, 53522, 56405,
                        59223, 62825, 66555, 69664, 72539, 75776, 75776, 75000, 74234],
                        name = 'stocks', 
                        index = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 
                                 2025, 2026, 2027, 2028, 2029, 2030, 2031, 2032, 2033, 2034, 
                                 2035, 2036, 2037, 2038, 2039, 2040, 2041, 2042, 2043, 2044, 
                                 2045, 2046, 2047, 2048, 2049, 2050]) # megawatts installed capacity of wind turbines

# survival curve parameters inputs and creation of the survival function
ages = range(0, len(input_data))  # the maximum age in the MFA model is equal to the length of the input data
mean = 20  # offshore turbines have a mean lifespan of 20 years (Fishman and Graedel 2019)
std_dev = 10  # for this example, changed from sd = 5 in Fishman and Graedel 2019.
survival_curve = scipy.stats.norm.sf(ages, loc=mean, scale=std_dev)

# material Intensites (MIs)
mi_Nd_high = 175.5  # material intensity of kg of Neodymium per MW in 2015.
mi_Nd_low  = 124    # material intensity of kg of Neodymium per MW in 2050.
# Linear interpolation of MIs over time, describing an assumed technological efficiency improvement from 2025 to 2050
mi_Nd = pd.Series(np.interp(input_data.index, [2015, 2025, 2050], [mi_Nd_high, mi_Nd_high, mi_Nd_low]),
                        name = 'mi', 
                        index = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024, 
                                 2025, 2026, 2027, 2028, 2029, 2030, 2031, 2032, 2033, 2034, 
                                 2035, 2036, 2037, 2038, 2039, 2040, 2041, 2042, 2043, 2044, 
                                 2045, 2046, 2047, 2048, 2049, 2050])

mi_Nd.plot(ylim=(0,200), ylabel="Nd/MW")
```

## 3. Set up the MFA systems

Set up two MFA systems, one for each of the two layers of products and materials, because they use different units of measurement.

```python
# %% 3. Set up MFA systems for the two layers of products and materials
units = {'time_interval':   'year', 
         'flows_products':  'MW/year', 
         'stocks_products': 'MW', 
         'flows_materials': 'Nd/year', 
         'stocks_materials':'Nd'}

timeframes = cohorts = input_data.index  # timeframes are the "snapshots" of the system in every time interval, while cohorts are the inflows that enter the system in every time interval. They are the same because we have one inflow per time interval.

time_interval = 1 # Interval between timeframes, usually 1, e.g. from 2005 to 2006

cohort_stop = (cohorts[-1] + time_interval) #because of exclusive slicing in python, we need to add an extra time step

mfa_system_products = pd.DataFrame(index = timeframes, columns = ['stocks',
                                                                  'inflows', 
                                                                  'outflows',
                                                                  'stock_changes'])

mfa_system_materials = pd.DataFrame(index = timeframes, columns = ['stocks',
                                                                   'inflows', 
                                                                   'outflows',
                                                                   'stock_changes'])

mfa_system_products['stocks'] = input_data
```

## 4. Create the survival curves matrix

In this example, the material (Nd) lasts as long as the product/service it's in (turbines in MW of energy generation) so the two layers share one survival curves matrix.

```python
# %% 4. Create survival curves matrix
# create survival curve matrix filled with zeros
survival_curves_matrix = pd.DataFrame(0, index=timeframes, columns=cohorts, dtype=float)

# and populate that matrix
for cohort in cohorts:
    cohort
    # at each iteration, we shift the survival curve and we fill the survival curve matrix with the shifted survival curve
    survival_curves_matrix.loc[cohort: , cohort] = survival_curve[0 : cohort_stop - cohort]
```

## 5. Calculate the MFA system of the products layer

In this example the stocks of the product (turbines in MW of energy generation) is given so this step is a stock-driven model, but can be replaced with a flow-driven model if inflows are the exogenous input data.

```python
# %% 5. Calculate the MFA system of the products layer

# Stock-driven model in this example.

# Create cohorts survival matrix and calculate the inflows
# initialize the cohort_survival_matrix filled with zeros
cohorts_survival_matrix_products = pd.DataFrame(0, index=timeframes, columns=cohorts, dtype=float)

# And populate it while creating each inflow
for cohort in cohorts:
    # calculate the inflow of the time-th year
    mfa_system_products.loc[cohort, 'inflows'] = (
        mfa_system_products.loc[cohort, 'stocks'] - cohorts_survival_matrix_products.loc[cohort, : cohort - 1].sum()
    ) / survival_curves_matrix.loc[cohort, cohort]
    # now that the time-th inflow is known, scale the shifted survival curve with this inflow
    cohorts_survival_matrix_products.loc[:, cohort] = survival_curves_matrix.loc[:, cohort] * mfa_system_products.loc[cohort, 'inflows']

# calculate stock changes (net change to stock)
# prepend=0 replicates the convention that stock_change is the difference between current timeframe and previous timeframe. 
mfa_system_products["stock_changes"] = np.diff(mfa_system_products["stocks"], prepend=0)

# Calculate outflows using mass balance
mfa_system_products["outflows"] = mfa_system_products["inflows"] - mfa_system_products["stock_changes"]
```

### View the full MFA system of the products layer

```python
# %% View the full MFA system of the products layer
# plot flows of products
mfa_system_products[["inflows", "outflows", "stock_changes"]].plot(ylabel=units['flows_products'])
```

```python
# plot stocks of products
mfa_system_products["stocks"].plot(kind="area", stacked=True, ylabel=units['stocks_products'])
```

## 6. Convert inflows from products to materials
Using the material intensities time series:
$
\textcolor{#2A8596}{\text{material inflows}(t)[mass]} = \textcolor{#C58D01}{\text{product inflows}(t)[units]} \times \textcolor{#C32D2E}{\text{material intensities}(t)[\frac{mass}{unit}]}
$

```python
# %% 6. Convert inflows from products to materials
# material inflows(t) = product inflows(t) x material intensities(t)

mfa_system_materials['inflows'] = mfa_system_products['inflows'] * mi_Nd 
```

## 7. Calculate the MFA system of the materials layer

Using a flow-driven MFA, because the inflows (in mass) are now available.

```python
# %% 7. Calculate the MFA system of the materials layer
# Using a flow-driven MFA, because the inflows (in mass) are now available.

# Create cohorts survival matrix
# initialize the cohort_survival_matrix filled with zeros
cohorts_survival_matrix_materials = pd.DataFrame(0, index=timeframes, columns=cohorts, dtype=float)
# And populate it
for cohort in cohorts:
    # scale the shifted survival curve with the inflow of the time-th timeframe
    cohorts_survival_matrix_materials.loc[:, cohort] = survival_curves_matrix.loc[:, cohort] * mfa_system_materials['inflows'].loc[cohort]

# the stocks are the sum of the surviving cohorts, which is the sum of cohorts survival matrix row by row.
mfa_system_materials["stocks"] = cohorts_survival_matrix_materials.sum(axis='columns')

# calculate stock changes (net change to stock)
# prepend=0 replicates the convention that stock_change is the difference between current timeframe and previous timeframe. 
mfa_system_materials["stock_changes"] = np.diff(mfa_system_materials["stocks"], prepend=0)

# Calculate outflows using mass balance
mfa_system_materials["outflows"] = mfa_system_materials["inflows"] - mfa_system_materials["stock_changes"]
```

### View the full MFA system of the materials layer

Compared with the equivalent plot of the products layer, not only have the units of measurement changed, __the trends are different__: inflows are flatter, stock changes start declining, etc.

```python
# %% View the full MFA system of the materials layer
# plot flows of materials
mfa_system_materials[["inflows", "outflows", "stock_changes"]].plot(ylabel=units['flows_materials'])
```

```python
# plot stocks of materials
mfa_system_materials["stocks"].plot(kind="area", stacked=True, ylabel=units['stocks_materials'])
```
